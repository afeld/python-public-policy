Retrieves NYC Open Data with a `Borough` column from the [Socrata Discovery API](https://dev.socrata.com/docs/other/discovery#?route=overview).


```python
import requests

response = requests.get(
    "https://data.cityofnewyork.us/api/catalog/v1",
    params={
        # # https://dev.socrata.com/docs/other/discovery#?route=get-/catalog/v1-column_names-name-
        "column_names": "Borough",
        # https://dev.socrata.com/docs/other/discovery#?route=get-/catalog/v1-limit-number-offset-number-
        "limit": 1000,
        # https://dev.socrata.com/docs/other/discovery#?route=get-/catalog/v1-only-type-
        "only": "dataset",
        # https://dev.socrata.com/docs/other/discovery#?route=get-/catalog/v1-order-sort_order-
        "order": "page_views_last_month DESC",
        # https://dev.socrata.com/docs/other/discovery#?route=get-/catalog/v1-provenance-provenance-
        "provenance": "official",
    },
)

# https://3.python-requests.org/user/quickstart/#json-response-content
response.raise_for_status()
```

```python
import pandas as pd

datasets = pd.json_normalize(response.json()["results"])
# datasets
```

```python
columns = ["resource.attribution", "resource.name", "link"]
datasets_simple = datasets[columns]
datasets_simple.style.format(hyperlinks="html")
```
