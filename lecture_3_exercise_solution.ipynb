```python
import pandas as pd

# load all records and force a consistent ordering
parks = pd.read_csv(
    "https://data.cityofnewyork.us/resource/enfh-gkve.csv?$limit=3000&$order=globalid"
)
parks
```

```python
import plotly.express as px

px.histogram(
    parks,
    x="borough",
    title="Number of parks per borough",
)
```
